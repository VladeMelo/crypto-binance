```python
import pandas as pd
```

```python
df = pd.read_csv('50-50 ETH LGBM Drawdown Stay Out (3 years) 15m.csv')
```

```python
df['profit_total'] = df['profit_1'] + df['profit_0']
df['x1'] = df['profit_1'] / (df['max_drawdown_1'] * -1)
df['x0'] = df['profit_0'] / (df['max_drawdown_0'] * -1)
df['x'] = df['x1'] + df['x0']
```
